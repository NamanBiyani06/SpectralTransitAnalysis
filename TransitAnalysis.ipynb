```python
from lightkurve import search_targetpixelfile
from lightkurve import TessTargetPixelFile
import lightkurve
import numpy as np
```

# Fetching Data using `lightkurve`

Analysis of Tabby's Star in order to familiarize with `lightkurve`

```python
pixelfile = search_targetpixelfile('KIC 6922244', author = "Kepler", cadence = "long", quarter = 4).download()
```

```python
# displaying 1 frame of the downloaded pixelfile
pixelfile.plot(frame = 1)
```

# Plotting and Analysing the Data

Creating the light curve from the pixel file data

```python
lc = pixelfile.to_lightcurve(aperture_mask = pixelfile.pipeline_mask)
lc.plot()
```

# Processing the Light Curve

Flattening the light curve to minimize the data into fewer dimensions

```python
flat_lc = lc.flatten()
flat_lc.plot()
```

Folding the light curve in order by stacking the spectral frequency spikes in order to reveal more information

```python
folded_lc = flat_lc.fold(period = 3.5225)
folded_lc.plot()
```
